```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train=data_Source.drop(['TotalOrgsWorked','Attrition','LastPromotionYears','AverageFeedback','LastPayIncrementBand'], axis=1)
test=data_Source['EmployeeID']
```

```python
X_train, X_test,Y_train, Y_test= train_test_split(train, test, test_size=0.4,random_state=2)
```

```python
regre=LinearRegression()
```

```python
regre.fit(X_train, Y_train)
```

```python
predct=regre.predict(X_test)
```

```python
predct
```

```python
regre.score(X_test, Y_test)
```
